##**Etapa 1: Preprocesamiento de los datos**

Aplicar metodos para reemplazar valores nulos y justificar el metodo aplicado por cada columna o dataframe


Necesitamos buscar un metodo para poder eliminar los valores nulos de las columnas con información cualitativa, por lo que en este caso yo opte por escoger el promedio de los datos

```python
#Aplicar metodos para reemplazar valores nulos y justificación
#Metodo cuantitativas
# Filtramos las columnas que contienen sólo valores enteros o flotantes es decir cuantitativas
cuantitativas1 = df1.select_dtypes(include=['int', 'float']).columns.tolist()
# Calculamos la media de cada columna numérica
column_means1 = df1[cuantitativas1].mean()
# Reemplazamos los valores nulos con el promedio
df1[cuantitativas1] = df1[cuantitativas1].fillna(column_means1)

# Filtramos las columnas que contienen sólo valores enteros o flotantes es decir cuantitativas
cuantitativas2 = df2.select_dtypes(include=['int', 'float']).columns.tolist()
# Calculamos la media de cada columna numérica
column_means2 = df2[cuantitativas2].mean()
# Reemplazamos los valores nulos con el promedio
df2[cuantitativas2] = df2[cuantitativas2].fillna(column_means2)

# Filtramos las columnas que contienen sólo valores enteros o flotantes es decir cuantitativas
cuantitativas3 = df3.select_dtypes(include=['int', 'float']).columns.tolist()
# Calculamos la media de cada columna numérica
column_means3 = df3[cuantitativas3].mean()
# Reemplazamos los valores nulos con el promedio
df3[cuantitativas3] = df3[cuantitativas3].fillna(column_means3)

display(df1[cuantitativas1].isnull().sum(),df2[cuantitativas2].isnull().sum(),df3[cuantitativas3].isnull().sum())
```

Ahora buscamos un metodo para poder eliminar los valores nulos de las columnas con información cuantitativa, en este caso opte por escoger el metodo forward fill, que toma el valor posterior de la columna, esto opte por hacerlo asi debido que es información no numerica y no puedo realizar otro metodo.

```python
#Sustituir valores nulos por valores no nulos hacia atrás backward fill" ("bfill")
df1 = df1.fillna(method='ffill')
df2 = df2.fillna(method='ffill')
df3 = df3.fillna(method='ffill')

df1 = df1.fillna(method='bfill')
df2 = df2.fillna(method='bfill')
df3 = df3.fillna(method='bfill')

#Corroboramos que ya no existen valores nulos
display(df1.isnull().sum().sum(),df2.isnull().sum().sum(),df3.isnull().sum().sum())
```

En efecto se puede observar que ya no existen valores nulos en las columnas.
